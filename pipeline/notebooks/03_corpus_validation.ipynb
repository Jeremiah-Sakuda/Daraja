This notebook validates the synthetic corpus using semantic similarity filtering with LaBSE embeddings.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / 'src'))

from data.validator import CorpusValidator, EmbeddingModel, run_validation_pipeline

SOURCE_LANG = "so"
TARGET_LANG = "sw"
SYNTHETIC_DIR = Path(f'../data/synthetic/{SOURCE_LANG}-{TARGET_LANG}')
OUTPUT_DIR = Path(f'../data/synthetic/{SOURCE_LANG}-{TARGET_LANG}/validated')
```

```python
# Load embedding model
embedding_model = EmbeddingModel("sentence-transformers/LaBSE")

# Initialize validator
validator = CorpusValidator(
    embedding_model=embedding_model,
    similarity_threshold=0.75,
    min_length=5,
    max_length=500
)
```

```python
# Run validation
report = run_validation_pipeline(
    source_file=str(SYNTHETIC_DIR / f'synthetic.{SOURCE_LANG}'),
    target_file=str(SYNTHETIC_DIR / f'synthetic.{TARGET_LANG}'),
    output_dir=str(OUTPUT_DIR),
    similarity_threshold=0.75
)
```

```python
print("=" * 50)
print("VALIDATION COMPLETE")
print("=" * 50)
print(f"\nPass rate: {report.pass_rate:.1%}")
print(f"Passed pairs: {report.passed_pairs:,}")
print(f"Rejected pairs: {report.rejected_pairs:,}")
print(f"\nNext step: Run 04_unsloth_finetuning.ipynb")
```
